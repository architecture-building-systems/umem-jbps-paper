```python
%matplotlib inline

import esoreader
reload(esoreader)
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import *
import csv
from __future__ import division
```

## Get Co-Simulation data

```python
# change this to point to where have the UMEM dropbox folder
PROJECT_PATH = r"c:\Dropbox\SuAT\UMEM\Quartier_Nord_EPFL"

# this is the output of the workflow
ESO_PATH = os.path.join(PROJECT_PATH,'cosimulation','quartier_nord.eso')

# yeah... we need an index for timeseries...
HOURS_IN_YEAR = pd.date_range('2013-01-01', '2013-12-31 T23:00', freq='H')
```

```python
qn_cosim_eplus = esoreader.read_from_path(ESO_PATH)
```

```python
qn_cosim_eplus.find_variable('Zone Windows Total Transmitted Solar Radiation Energy')
```

```python
qn_cosim_eplus?
```

```python
solar_gains = qn_cosim_eplus.to_frame('Zone Windows Total Transmitted Solar Radiation Energy', index=HOURS_IN_YEAR)
solar_gains.resample('W', how='sum').plot()
```

## Let us look at two days (cooling and heating case)

cooling case

```python
startdate, enddate = ('2013-07-01','2013-07-31')
sgcool = solar_gains.truncate(before=startdate, after=enddate).resample('2H', how='mean')
sgcool['Time'] = sgcool.index.map(lambda t: t.time())
sgcool.groupby(by='Time').mean().plot()
```

heating case

```python
startdate, enddate = ('2013-01-01','2013-01-31')
sgheat = solar_gains.truncate(before=startdate, after=enddate).resample('2H', how='mean')
sgheat['Time'] = sgheat.index.map(lambda t: t.time())
sgheat.groupby(by='Time').mean().plot()
```

# CitySim data

```python
COLUMN = '1(1):1:Qi(Wh)'

cs_df = pd.read_table(os.path.join(PROJECT_PATH, "nord_only_CitySim", "nord_only_CitySim_TH.out"))
cs_df.index = HOURS_IN_YEAR
qi_df = pd.DataFrame({'Qi': cs_df[COLUMN]})

# convert Qi from [Wh] to [J]
qi_df['Qi'] *= 3600
```

```python
qi_df.resample('W', how='sum').plot()
```

```python
startdate, enddate = ('2013-07-01','2013-07-31')
sgcool = qi_df.truncate(before=startdate, after=enddate).resample('2H', how='mean')
sgcool['Time'] = sgcool.index.map(lambda t: t.time())
sgcool.groupby(by='Time').mean().plot()
```

```python
startdate, enddate = ('2013-01-01','2013-01-31')
sgheat = qi_df.truncate(before=startdate, after=enddate).resample('2H', how='mean')
sgheat['Time'] = sgheat.index.map(lambda t: t.time())
sgheat.groupby(by='Time').mean().plot()
```

# Compare CitySim solo and cosim

```python
cosim = solar_gains.sum(axis=1)
df = pd.DataFrame({'cosim': cosim, 'solo': qi_df['Qi']})
df.resample('W', how='sum').plot()                   
```

```python
startdate, enddate = ('2013-07-01','2013-07-31')
sgcool = df.truncate(before=startdate, after=enddate).resample('2H', how='mean')
sgcool['Time'] = sgcool.index.map(lambda t: t.time())
sgcool.groupby(by='Time').mean().plot()
```

```python
startdate, enddate = ('2013-01-01','2013-01-31')
sgheat = df.truncate(before=startdate, after=enddate).resample('2H', how='mean')
sgheat['Time'] = sgheat.index.map(lambda t: t.time())
sgheat.groupby(by='Time').mean().plot()
```
