# let's do Spearman

```python
Indicators_2012 = pd.read_csv('/Users/Clayton/Dropbox/WDIData/WDI extract_2012.csv')
```

```python
Indicators_2012_nona = Indicators_2012[['name','Unnamed: 2','BN.KLT.DINV.CD']].dropna()
```

```python
Indicators_2012_nona.info()
```

```python
import scipy.stats as stats
```

```python
results = {}
for columnname in Indicators_2012.columns[4:]:
    try:
        print columnname
        df = Indicators_2012[['name','Unnamed: 2', columnname]].dropna()
        #print stats.spearmanr(df['Unnamed: 2'], df[columnname])
        results[columnname] = stats.spearmanr(df['Unnamed: 2'], df[columnname])
    except:
        print "can't compute"+columnname
```

```python
results['DT.DOD.MLAT.ZS'][0]
```

```python
results = pd.DataFrame(results)
```

```python
results = results.T
```

```python
results.columns = ['Corr','Pval']
```

```python
results.sort_values('Corr', ascending=False)
```

```python
mergedresults = pd.merge(results, Indicators_2012.ix[:0].T.dropna(), right_index=True, left_index=True)
```

```python
mergedresults.sort_values('Corr', ascending=False).to_csv('/Users/Clayton/Dropbox/WDIData/Spearman_2012.csv')
```
